```python
data=food_choices.copy() #copying data for modification, do not use real data for modification
```

```python
#remove white spaces from columns
data.columns = data.columns.str.strip() #converts into strings and forms as lists
data.columns.tolist()
```

```python
# Removing unwanted data in place
data.drop(['GPA','Gender', 'comfort_food_reasons_coded.1', 'eating_changes_coded1','employment','father_education',
'father_profession', 'grade_level','income','mother_profession','mother_education', 'ethnic_food', 'on_off_campus','sports', 
'type_sports', 'marital_status'], axis = 1, inplace = True)
data.columns.tolist()
```

```python
#remove code that doesn't have any 
data.dropna(subset=['calories_day','weight'], inplace=True)
data.isna().sum()
```

```python
# Remove rows where user didn't have any important details
missing_data = data[(data['self_perception_weight'].isna()) & (data['nutritional_check'].isna())]
data.drop(missing_data.index, inplace=True)

# Remove favorite cusinie column
data.drop(columns=['fav_cuisine'], axis=1, inplace=True)
```

```python
data.drop(['vitamins'],axis = 1, inplace = True)
data.head(10)
```

```python
df = pd.DataFrame(columns= ['weight', 'nutritional_check', 'healthy_meal','comfort_food_reasons', 'eating_changes','diet_current','veggies_day','waffle_calories'])
df['weight'] = data['weight']
df['nutritional_check'] = data['nutritional_check']
df['waffle_calories'] = data['waffle_calories']
df['healthy_meal'] = data['healthy_meal']
df['comfort_food_reasons'] = data['comfort_food_reasons']
df['eating_changes'] = data['eating_changes']
df['diet_current'] = data['diet_current']
df['veggies_day'] = data['veggies_day']
df.head()
#creating a dataframe and keeping only the required data
```

```python
df = df.drop_duplicates() # remove duplicates
```

```python
print(df.isnull().sum())  # check for missing values
# drop rows with any missing values
df = df.dropna()
```

```python
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df['waffle_calories'] = pd.to_numeric(df['waffle_calories'], errors='coerce')
df['veggies_day'] = pd.to_numeric(df['veggies_day'], errors='coerce')
df.head(10)
#converting into similar datatype
```

```python
df['healthy_meal'] = df['healthy_meal'].str.strip().str.lower()
df['comfort_food_reasons'] = df['comfort_food_reasons'].str.strip().str.lower()
df['eating_changes'] = df['eating_changes'].str.strip().str.lower()
df['diet_current'] = df['diet_current'].str.strip().str.lower()
#standardize text
```
